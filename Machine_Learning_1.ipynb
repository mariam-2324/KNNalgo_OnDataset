#Importing the Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

#Loading the Dataset from the SKlearn Lib

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
print(iris['feature_names'])
print(iris['target_names'])
print(iris['data'][0:5])
print(iris['target'][0:5])
```

```python


# Create DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

df.head()
```

#Data Visualization


```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species', corner=True)
plt.suptitle("Iris Dataset Visualization", y=1.02)
plt.show()
```

```python
iris["data"]
```

# Spliting the Data into TEST and TRAINING

```python
from sklearn.model_selection import train_test_split

X = iris.data #questions
y = iris.target  #Answers

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

#Training The DATA

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=7)  # k=5
KNN.fit(X_train, y_train)
```

#TESTING

```python
sample = [[5, 2.9, 1, 0.2]]  # petal length, petal width
pred = KNN.predict(sample)
print("Predicted species:", iris.target_names[pred][0])
print(pred)
```

#Testing Through our TEST DATA

```python
print(X_test)
```

```python
X_test.shape
```

```python
y_pred = KNN.predict(X_test)
print(f"Test set predictions:\n {y_pred}")
```

```python
print(y_test)
print(y_pred)
```

#ACCURACY


```python
print(f"Test set score: {np.mean(y_pred == y_test):.2f}")
```

```python
print(f"Test set score: {KNN.score(X_test, y_test):.2f}")
```

```python
from sklearn.metrics import accuracy_score
# train_acc = accuracy_score(y_train, KNN.predict(X_train))
test_acc = accuracy_score(y_test, KNN.predict(X_test))

# print(f"Train Accuracy: {train_acc*100:.2f}%")
print(f"Test Accuracy: {test_acc*100:.2f}%")
```

```python

# Select two features for visualization
X_2_features = X[:, [2, 3]]  # Using petal length and petal width

# Create a meshgrid for plotting the decision boundary
x_min, x_max = X_2_features[:, 0].min() - 1, X_2_features[:, 0].max() + 1
y_min, y_max = X_2_features[:, 1].min() - 1, X_2_features[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))


for k in [11, 9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_2_features, y)  # Train the model with 2 features
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 5))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    sns.scatterplot(x=X_2_features[:, 0], y=X_2_features[:, 1], hue=iris.target_names[y], s=50)
    plt.title(f'KNN Decision Boundary (k={k})')
    plt.xlabel('Petal length (cm)')
    plt.ylabel('Petal width (cm)')
    plt.show()
```
